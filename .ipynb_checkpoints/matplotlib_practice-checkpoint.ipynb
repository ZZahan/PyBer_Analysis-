```python
#Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig =plt.figure()
ax=fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a lable for the legend.

plt.plot(x_axis, y_axis, label ='Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")

plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim (0, 45)

#Create a title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()
```

```python
# Create the plot and add a lable for the legend.

plt.plot(x_axis, y_axis, marker="^", color ="green", linewidth = 2, label ='city of Boston')

# Create labels for the x and y axes.

plt.xlabel("Date")

plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim (0, 45)

#Create a title.

plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()


# Add the legend.

plt.legend()
```

```python
import matplotlib.pyplot as plt
```

```python
categories = ['pant', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
```

```python
revenue = [25, 15, 5, 20, 10, 25]
```

```python
plt.pie(revenue)
plt.show()
```

```python
plt.pie(revenue, labels=categories)
plt.show()
```

```python
plt.pie(revenue, labels=categories, autopct="%1.1f%%")
plt.show()
```

```python
explode= [0,1,0,0,0,0]
```

```python
plt.pie(revenue, labels=categories, autopct="%1.1f%%", explode=explode)
plt.show()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 1 colors, on for each month.
colors= ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "mistyrose"]
explode_values= (0,0,0,0,0,0,0.2,0,0,0,0,0,)
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')
plt.show()
```
